## Part 1: Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
import pandas as pd
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras import layers
#from tensorflow.keras.layers import Input, Dense



#  Import and read the attrition data
attrition_df = pd.read_csv('https://static.bc-edx.com/ai/ail-v-1-0/m19/lms/datasets/attrition.csv')
attrition_df.head()
```

```python
# Determine the number of unique values in each column.
attrition_df.nunique()
```

```python
# Create y_df with the Attrition and Department columns
y_df = attrition_df[['Attrition', 'Department']]
y_df.head()
```

```python
# Create a list of at least 10 column names to use as X data

# Select 10 columns for X (excluding Attrition and Department)
selected_columns = ['Age', 'DistanceFromHome', 'Education', 'JobLevel', 'HourlyRate',
                    'NumCompaniesWorked', 'PercentSalaryHike', 'TotalWorkingYears',
                    'YearsAtCompany', 'YearsSinceLastPromotion']

# Create X_df using your selected columns
X_df = attrition_df[selected_columns]
#X_df.head()

# Show the data types for X_df
X_df.dtypes
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_df, y_df, test_size=0.2, random_state=42)

# Display the shapes to confirm split
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)

```

```python
# Convert your X data to numeric data types however you see fit
# Add new code cells as necessary
```

```python
# Create a StandardScaler
scaler = StandardScaler()

# Fit the StandardScaler to the training data
X_train_scaled = scaler.fit_transform(X_train)


# Scale the training and testing data

# Scale the testing data
X_test_scaled = scaler.transform(X_test)


# Convert back to DataFrame for easier handling
X_train_scaled = pd.DataFrame(X_train_scaled, columns=selected_columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=selected_columns)

X_train_scaled.head()

```

```python
# Create a OneHotEncoder for the Department column
dept_encoder = OneHotEncoder(sparse=False)


# Fit the encoder to the training data
dept_encoded_train = dept_encoder.fit_transform(y_train[['Department']])

# Transform the department column in testing data
dept_encoded_test = dept_encoder.transform(y_test[['Department']])


# Create two new variables by applying the encoder
# to the training and testing data
# Convert the results to DataFrames for easier handling
dept_encoded_train_df = pd.DataFrame(dept_encoded_train, columns=dept_encoder.get_feature_names_out(['Department']))
dept_encoded_test_df = pd.DataFrame(dept_encoded_test, columns=dept_encoder.get_feature_names_out(['Department']))

dept_encoded_train_df.head()

```

```python
# Create a OneHotEncoder for the Attrition column
attrition_encoder = OneHotEncoder(sparse=False)


# Fit the encoder to the training data
attrition_encoded_train = attrition_encoder.fit_transform(y_train[['Attrition']])


# Create two new variables by applying the encoder
# to the training and testing data

# Transform the attrition column in testing data
attrition_encoded_test = attrition_encoder.transform(y_test[['Attrition']])

# Convert the results to DataFrames for easier handling
attrition_encoded_train_df = pd.DataFrame(attrition_encoded_train, columns=attrition_encoder.get_feature_names_out(['Attrition']))
attrition_encoded_test_df = pd.DataFrame(attrition_encoded_test, columns=attrition_encoder.get_feature_names_out(['Attrition']))

attrition_encoded_train_df.head()

```

## Create, Compile, and Train the Model

```python
# Find the number of columns in the X training data
input_dim = X_train_scaled.shape[1]
print(f'Number of input features: {input_dim}')


# Create the input layer
input_layer = Input(shape=(input_dim,), name='input')


# Create at least two shared layers

# Create the first shared layer with 64 units
shared_layer_1 = Dense(units=64, activation='relu', name='shared1')(input_layer)

# Create the second shared layer with 128 units
shared_layer_2 = Dense(units=128, activation='relu', name='shared2')(shared_layer_1)
```

```python
# Create a branch for Department
# with a hidden layer and an output layer


# Create the hidden layer
dept_hidden_layer = Dense(units=32, activation='relu', name='department_hidden')(shared_layer_2)


# Create the output layer
dept_output_layer = Dense(units=3, activation='softmax', name='department_output')(dept_hidden_layer)

```

```python
# Create a branch for Attrition
# with a hidden layer and an output layer

# Create the hidden layer
attrition_hidden_layer = Dense(units=32, activation='relu', name='attrition_hidden')(shared_layer_2)


# Create the output layer
attrition_output_layer = Dense(units=2, activation='softmax', name='attrition_output')(attrition_hidden_layer)

```

```python
# Create the model
model = Model(inputs=input_layer, outputs=[dept_output_layer, attrition_output_layer], name='model')


# Compile the model
model.compile(optimizer='adam',
              loss={'department_output': 'categorical_crossentropy', 'attrition_output': 'categorical_crossentropy'},
              metrics={'department_output': 'accuracy', 'attrition_output': 'accuracy'})


# Summarize the model
model.summary()
```

```python
# Train the model
history = model.fit(X_train_scaled,
                    {'department_output': dept_encoded_train_df, 'attrition_output': attrition_encoded_train_df},
                    epochs=100,
                    batch_size=10,
                    validation_split=0.2)

```

```python
# Evaluate the model with the testing data
evaluation = model.evaluate(X_test_scaled,
                            {'department_output': dept_encoded_test_df, 'attrition_output': attrition_encoded_test_df})

# Print the evaluation list to understand its structure
print("Evaluation List:", evaluation)

# Adjust the print statements based on the available elements
if len(evaluation) == 3:
    print(f"Overall Loss: {evaluation[0]}")
    print(f"Department Predictions Accuracy: {evaluation[1]}")
    print(f"Attrition Predictions Accuracy: {evaluation[2]}")
else:
    print("Unexpected evaluation output structure.")
```
